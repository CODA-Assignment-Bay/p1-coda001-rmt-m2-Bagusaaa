### Apa produk yang paling laris terjual?

````python
# Membuat kolom baru untuk tahun dan bulan
df['year'] = df['invoice_date'].dt.year
df['month'] = df['invoice_date'].dt.month

# Filter data untuk tahun 2020 dan 2021
data_2020 = df[df['year'] == 2020]
data_2021 = df[df['year'] == 2021]

# Hitung jumlah penjualan setiap produk untuk masing-masing tahun
sales_2020 = data_2020.groupby('product')['units_sold'].mean().reset_index()
sales_2021 = data_2021.groupby('product')['units_sold'].mean().reset_index()

# Pilih produk teratas untuk masing-masing tahun
top_product_2020 = sales_2020.sort_values(by='units_sold', ascending=False)
top_product_2021 = sales_2021.sort_values(by='units_sold', ascending=False)

# Buat visualisasi untuk produk teratas pada tahun 2020 dan 2021
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.bar(top_product_2020['product'], top_product_2020['units_sold'], color='skyblue')
plt.title('Produk Terlaris Tahun 2020')
# plt.xlabel('Produk')
plt.ylabel('Jumlah Penjualan')
plt.xticks(rotation=45, ha='right')

plt.subplot(1, 2, 2)
plt.bar(top_product_2021['product'], top_product_2021['units_sold'], color='Orange')
plt.title('Produk Terlaris Tahun 2021')
# plt.xlabel('Produk')
plt.ylabel('Jumlah Penjualan')

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

### Bagaimana trend penjualan produk nike di tahun 2020 - 2021 setiap bulannya?

menganalisis kinerja penjualan produk nike di tahun 2020 dan 2021. Penjualan yang dimaksud adalah pendapatan bulanan dari kedua tahun tersebut. Tujuan analisis ini adalah untuk mencaritahu seberapa banyak penjualan di 2 tahun tersebut dan untuk menentukan strategry apakah yang bakalan di pake untuk kedepanya.

````python
# Membuat kolom baru untuk tahun dan bulan
df['year'] = df['invoice_date'].dt.year
df['month'] = df['invoice_date'].dt.month

# Filter data untuk tahun 2020 dan 2021
data_2020 = df[df['year'] == 2020]
data_2021 = df[df['year'] == 2021]

# Menghitung total penjualan per bulan untuk tahun 2020 dan 2021
total_sales_2020 = data_2020.groupby('month')['total_sales'].mean()
total_sales_2021 = data_2021.groupby('month')['total_sales'].mean()

# Membuat line chart untuk trend penjualan per bulan
plt.figure(figsize=(10, 6))
plt.plot(total_sales_2020.index, total_sales_2020.values, label='2020', marker='o')
plt.plot(total_sales_2021.index, total_sales_2021.values, label='2021', marker='o')

# Menambahkan judul
plt.title('Trend Penjualan per Bulan (2020 vs 2021)')

# Menambahkan label tiap axis
plt.xlabel('Bulan')
plt.ylabel('Rata-Rata Penjualan')

# Menambahkan legend
plt.legend()
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'Mei', 'Jun', 'Jul', 'Agu', 'Sep', 'Okt', 'Nov', 'Des'])

# tampilkan visualisasi
plt.show()
````

````python
# Kelompokkan data berdasarkan tahun dan hitung total penjualan
total_sales_by_year = df.groupby('year')['total_sales'].sum()

# Menghitung persentase penjualan
sales_2020 = total_sales_by_year.loc[2020]
sales_2021 = total_sales_by_year.loc[2021]
percentage_2020 = (sales_2020 / (sales_2020 + sales_2021)) * 100
percentage_2021 = (sales_2021 / (sales_2020 + sales_2021)) * 100

# Plotting bar chart
plt.figure(figsize=(8, 6))
plt.bar(['2020', '2021'], [percentage_2020, percentage_2021], color=['blue', 'orange'])
plt.xlabel('Tahun')
plt.ylabel('Persentase Penjualan (%)')
plt.title('Persentase Penjualan antara Tahun 2020 dan 2021')
plt.ylim(0, 100)

# Menambahkan label pada tiap bar
for i, v in enumerate([percentage_2020, percentage_2021]):
    plt.text(i, v + 2, f'{v:.2f}%', ha='center')

plt.show()
````

### Apakah ada perbedaan yang signifikan antara penjualan dengan sales method offline dan online? inversial 

````python
df.sales_method.unique()
````

````python
df.info()
````

````python
pl = df
pl['sales_method'] = pl['sales_method'].replace({'Online':1,'In-store':0, 'Outlet':0})
````

````python
pl['sales_method'] = pl.sales_method.astype('int')
````

````python
#Extract groups
online = pl[pl['sales_method'] == 1]['total_sales']
offline = pl[pl['sales_method'] == 0]['total_sales']


t_stat_two_sample, p_value_two_sample =  stats.ttest_ind(online.sample(40), offline.sample(40))#we take 40 sample of each data to make the size the same.

#Perform dependent t-test and p-value
print(f"\nT-statistic: {t_stat_two_sample:.2f}")
print(f"P-value: {p_value_two_sample:.4f}")

#Conclusion
if p_value_two_sample < 0.05: 
    print("Fail to reject the null hypothesis. Tidak ada perbedaan penjualan yang signifikan antara online dan offline.")
else:
    print("Reject the null hypothesis. Terdapat perbedaan penjualan yang signifikan antara online dan offline.")
````

berdasarkan hasil uji hipotesis, dengan menggunakan *two sample test* menunjukan P-value: **0.0156**, dengan kata lain gagal menolak h0, ini menunjukan bahwa tidak ada perbedaan penjualan yang signifikan antara online dan Bukan offline.

### Apakah semakin murah harga barang mempengaruhi penjualan?

````python
#Computing Spearman Correlation Test
corr = df['price_per_unit'].corr(df['units_sold'], method='spearman')

print(f"\nPearson Correlation: {corr:.2f}")
sns.lmplot(data=df,x='price_per_unit', y='units_sold')
````

### Apakah ada perbedaan rata-rata Total Sales di berbagai Region (Wilayah)? 

````python
df.region.unique()
````

````python
Northeast = df[df['region']=='Northeast']
South = df[df['region']=='South']
West = df[df['region']=='West']
Midwest = df[df['region']=='Midwest']
Southeast = df[df['region']=='Southeast']
````

````python
res = stats.f_oneway(Northeast['total_sales'], South['total_sales'] , West['total_sales'], Midwest['total_sales'], Southeast['total_sales'])
print('P-value:',res.pvalue)
````

H0: Tidak terdapat perbedaan yang signifikan antara total penjualan antar wilayah

H1: Terdapat perbedaan yang signifikan antara total penjualan antar wilayah

```Hasil Uji```

Berdasarkan hasil uji, diperoleh nilai p-value kurang dari 0,05. Hal ini mengindikasikan bahwa hipotesis nol (H0) ditolak. Dengan demikian, hipotesis alternatif (H1) diterima, yang berarti terdapat perbedaan yang signifikan pada total penjualan antar wilayah.
